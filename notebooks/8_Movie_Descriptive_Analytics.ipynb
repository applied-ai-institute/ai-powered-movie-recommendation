## Step 12: Analyze Multi-Genre Movies

Find movies with multiple genres.

```python
# Count genres per movie
movies_df['genre_count'] = movies_df['genres'].apply(lambda x: len(x.split('|')))

print("\n" + "=" * 70)
print("MULTI-GENRE ANALYSIS")
print("=" * 70)

genre_count_dist = movies_df['genre_count'].value_counts().sort_index()

print(f"\nDistribution of genre count per movie:")
for count, num_movies in genre_count_dist.items():
    print(f"  {count} genre(s): {num_movies:,} movies")

print(f"\nAverage genres per movie: {movies_df['genre_count'].mean():.2f}")
print(f"Most genres in a single movie: {movies_df['genre_count'].max()}")

# Show examples of movies with most genres
max_genres = movies_df['genre_count'].max()
print(f"\n{'='*70}")
print(f"MOVIES WITH {max_genres} GENRES")
print("=" * 70)

multi_genre_movies = movies_df[movies_df['genre_count'] == max_genres][['title', 'genres']].head(5)
display(multi_genre_movies)
```

## Step 13: Create Comprehensive Excel Report

Export all analysis results to a multi-sheet Excel file.

```python
# Create comprehensive Excel report
excel_report = os.path.join(output_dir, "movielens_genre_analysis.xlsx")

with pd.ExcelWriter(excel_report, engine='openpyxl') as writer:
    # Sheet 1: Genre Counts (sorted by count)
    genre_counts.to_excel(writer, sheet_name='Genre Counts', index=False)
    
    # Sheet 2: Summary Statistics
    summary_df = pd.DataFrame({
        'Metric': list(summary_stats.keys()),
        'Value': list(summary_stats.values())
    })
    summary_df.to_excel(writer, sheet_name='Summary Statistics', index=False)
    
    # Sheet 3: Genre Distribution per Movie
    genre_dist_df = pd.DataFrame({
        'Number of Genres': genre_count_dist.index,
        'Number of Movies': genre_count_dist.values
    })
    genre_dist_df.to_excel(writer, sheet_name='Genres per Movie', index=False)
    
    # Sheet 4: Sample Movies (first 100)
    movies_df.head(100).to_excel(writer, sheet_name='Sample Movies', index=False)
    
    # Adjust column widths for all sheets
    for sheet_name in writer.sheets:
        worksheet = writer.sheets[sheet_name]
        for column in worksheet.columns:
            max_length = 0
            column_letter = column[0].column_letter
            for cell in column:
                try:
                    if len(str(cell.value)) > max_length:
                        max_length = len(cell.value)
                except:
                    pass
            adjusted_width = min(max_length + 2, 50)
            worksheet.column_dimensions[column_letter].width = adjusted_width

print(f"\n✅ Comprehensive Excel report created: {excel_report}")
print(f"   Sheets included:")
print(f"   1. Genre Counts - All {len(genre_counts)} genres with counts")
print(f"   2. Summary Statistics - Key metrics")
print(f"   3. Genres per Movie - Distribution analysis")
print(f"   4. Sample Movies - First 100 movies from dataset")
```
