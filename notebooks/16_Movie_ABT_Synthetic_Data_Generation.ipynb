### Import Librares

```python
import numpy as np
import pandas as pd

```

### Input Example Data

```python
# Define the variables and the target correlation matrix
variables = [
    'avg_rating', 'number_of_viewers', 'budget', 'popularity',
    'revenue', 'vote_average', 'vote_count'
]
```

```python
# The correlation matrix
# Note: Since it's symmetric, we can input it as a list of lists or a 2D array.
correlation_matrix = np.array([
    [1.000, 0.127, -0.092, 0.061, 0.003, 0.531, 0.099],
    [0.127, 1.000, 0.285, 0.551, 0.441, 0.284, 0.623],
    [-0.092, 0.285, 1.000, 0.520, 0.763, 0.001, 0.620],
    [0.061, 0.551, 0.520, 1.000, 0.636, 0.261, 0.794],
    [0.003, 0.441, 0.763, 0.636, 1.000, 0.140, 0.782],
    [0.531, 0.284, 0.001, 0.261, 0.140, 1.000, 0.299],
    [0.099, 0.623, 0.620, 0.794, 0.782, 0.299, 1.000]
])
```

### Setup Synthetic Inputs

```python
# For synthetic data generation, we assume the variables are standardized (mean=0, std=1).
# In this case, the covariance matrix IS the correlation matrix.
covariance_matrix = correlation_matrix
```

```python
# Number of rows (samples) to generate
num_samples = 500

# Mean vector (assuming standardized variables)
mean_vector = np.zeros(len(variables))
```

### --- Method 1: Using numpy.random.multivariate_normal (Direct Approach) ---

```python
print("--- Generating Data with Method 1 (multivariate_normal) ---")

# The function handles the transformation process internally
synthetic_data_direct = np.random.multivariate_normal(
    mean=mean_vector,
    cov=covariance_matrix,
    size=num_samples
)

# Convert to a Pandas DataFrame for better visualization
df_direct = pd.DataFrame(synthetic_data_direct, columns=variables)

print("\nFirst 10 rows of generated data (direct approach):")
print(df_direct.head(10))



```

```python
# Verify the result
corr_m1 = df_direct.corr()

# print("\nCalculated Correlation Matrix (should be close to target):")
print(corr_m1)
```

### --- Method 2: Using Cholesky Decomposition (Illustrative Approach) ---

```python

print("\n--- Generating Data with Method 2 (Cholesky Decomposition) ---")

# Step 1: Generate uncorrelated (white noise) standard normal random data
uncorrelated_data = np.random.randn(num_samples, len(variables))
uncorrelated_data
```

```python
# 

# Step 2: Calculate the Cholesky decomposition of the target covariance matrix
# This finds a lower triangular matrix 'L' such that L * L^T = covariance_matrix
try:
    L = np.linalg.cholesky(covariance_matrix)
    print("\nCholesky decomposition (Lower triangular matrix L):")
    print(L)
except np.linalg.LinAlgError:
    print("\nError: The provided covariance matrix is not positive-definite. Cholesky decomposition failed.")
    # Exit or use a different generation method if this fails
    exit()

```

```python
# Step 3: Transform the uncorrelated data using the decomposition matrix L
# The dot product applies the desired correlation structure
# 
correlated_data = uncorrelated_data.dot(L.T)
correlated_data

```

```python
# Step 4: Add the mean vector (assuming 0 mean for this example)
synthetic_data_cholesky = correlated_data + mean_vector
synthetic_data_cholesky

```

```python
# Convert to a Pandas DataFrame
df_cholesky = pd.DataFrame(synthetic_data_cholesky, columns=variables)
print("\nFirst 10 rows of generated data (Cholesky approach):")
print(df_cholesky.head(10))
```

```python
# Verify the result
corr_m2 = df_cholesky.corr()

# print("\nCalculated Correlation Matrix (should be close to target):")
print(corr_m2)

```

```python
# --- ACCURACY CALCULATION (Mean Absolute Error) ---
# We calculate the average difference between the Target and Synthetic cells
mae_m1 = np.mean(np.abs(correlation_matrix - corr_m1))
mae_m2 = np.mean(np.abs(correlation_matrix - corr_m2))

print(f"Comparison Results (N={num_samples})")
print("-" * 35)
print(f"Method 1 MAE: {mae_m1:.6f}")
print(f"Method 2 MAE: {mae_m2:.6f}")
print("-" * 35)
print(f"Difference: {abs(mae_m1 - mae_m2):.6f}")
```
